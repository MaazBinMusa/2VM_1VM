```python
import os
import json
import csv
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = []
count = {}

with open('run8_2021-02-23_final.json','r') as file:
    data = json.load(file)

for item in data:
    url = item['img_url']
    hash_ = item['img_hash']

    cat = url.split("_")[3]
    num = int(url.split("_")[2])
    
    if(num not in count):
        count[num] = {}

    if(cat not in count[num]):
        count[num][cat] = set()
        #count[num][cat] = []

    count[num][cat].add(hash_)
    #count[num][cat].append(hash_)
```
